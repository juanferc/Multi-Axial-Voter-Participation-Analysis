```python
import pandas as pd
```

```python
projFile = "Florida_educational_attainment_voting_registration_income.csv"
dfProj = pd.read_csv(projFile)
dfProj
```

```python
dfGeo = pd.read_csv("us-zip-code-latitude-and-longitude.csv", sep=";")
dfGeo
```

```python
dfFnl = pd.merge(dfProj, dfGeo, left_on= 'Zip_CoDE', right_on = 'Zip')
```

```python
dfProj.columns
```

```python
dfFnl1 = dfFnl[['Zip_CoDE', 'less_than_grade_9_est', 'high_school_non_grad_est',
       'high_school_grad_est', 'some_college_no_degree_est',
       'associates_degree_est', 'bachelor_degree_est',
       'grad_or_professional_degree_est', 'registration_rate', 'turnout_rate',
       'latitude', 'longitude', 'population', 'num_registered_voters',
       'num_voted', 'did_not_vote', 'average_house_value', 'average_income','Latitude','Longitude','City','State']]
```

```python
dfFnl1['Longitude'].isna().sum()
```

```python
dfFnl1.to_csv("FinalProjectData.csv", index = False)
```

```python
map_geojson = {
  "type": "FeatureCollection",
  "features" : []
}
```

```python
def genFeatureDict(info):
    f = {
        "type" : "Feature",
        "geometry":{
            "type": "Point",
            "coordinates": [info.Latitude,info.Longitude]
        },
        "properties" : {
            "City" : info.City,
            "zipcode" : info.Zip_CoDE,
            "Turnout Rate" : info.turnout_rate
        },

    }
    return f
```

```python
for r in dfFnl1.iterrows():
#     print(r[1])
    map_geojson['features'].append(genFeatureDict(r[1]))
```
